```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate sample data (you can use your own dataset)
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

# Create KMeans instance with the desired number of clusters (k)
kmeans = KMeans(n_clusters=4)

# Fit the model to the data
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualize the clustered data and centroids
plt.figure(figsize=(8, 6))

# Plot data points with their assigned clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7, edgecolors='w')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='o', s=200, edgecolors='k', label='Centroids')
plt.title('KMeans Clustering')
plt.legend()
plt.show()
```
